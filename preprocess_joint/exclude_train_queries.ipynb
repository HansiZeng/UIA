```python
import os 
import ujson 

in_dir = "/home/jupyter/jointly_rec_and_search/datasets/jointly_rec_and_search/"
out_dir = "/home/jupyter/jointly_rec_and_search/datasets/jointly_rec_and_search/exclude_train_queries/"
if not os.path.exists(out_dir):
    os.mkdir(out_dir)

# anchor
train_anchors = set()
with open(os.path.join(in_dir, "train_anchors.tsv")) as fin:
    for line in fin:
        anchor_id, _ = line.strip().split("\t")  
        train_anchors.add(anchor_id)

# --------
val_anchor_in_train = 0
with open(os.path.join(in_dir, "val_anchors.tsv")) as fin:
    with open(os.path.join(out_dir, "val_anchors.tsv"), "w")  as fout:
        for line in fin:
            anchor_id, _ = line.strip().split("\t")
            if anchor_id in train_anchors:
                val_anchor_in_train += 1
                continue
            fout.write(line)

with open(os.path.join(in_dir, "val_anchor_qrels.tsv")) as fin:
    with open(os.path.join(out_dir, "val_anchor_qrels.tsv"), "w") as fout:
        for line in fin:
            anchor_id, _, _, _ = line.strip().split("\t")
            if anchor_id in train_anchors:
                continue
            fout.write(line)
            
with open(os.path.join(in_dir, "val_anchor_qrels_clicknum.tsv")) as fin:
    with open(os.path.join(out_dir, "val_anchor_qrels_clicknum.tsv"), "w") as fout:
        for line in fin:
            anchor_id, _, _ = line.strip().split("\t")
            if anchor_id in train_anchors:
                continue
            fout.write(line)

# -------------
test_anchor_in_train = 0
with open(os.path.join(in_dir, "test_anchors.tsv")) as fin:
    with open(os.path.join(out_dir, "test_anchors.tsv"), "w")  as fout:
        for line in fin:
            anchor_id, text = line.strip().split("\t")
            if anchor_id in train_anchors:
                test_anchor_in_train += 1
                continue
            fout.write(line)

with open(os.path.join(in_dir, "test_anchor_qrels.tsv")) as fin:
    with open(os.path.join(out_dir, "test_anchor_qrels.tsv"), "w") as fout:
        for line in fin:
            anchor_id, _, _, _ = line.strip().split("\t")
            if anchor_id in train_anchors:
                continue
            fout.write(line)
            
with open(os.path.join(in_dir, "test_anchor_qrels_clicknum.tsv")) as fin:
    with open(os.path.join(out_dir, "test_anchor_qrels_clicknum.tsv"), "w") as fout:
        for line in fin:
            anchor_id, _, _ = line.strip().split("\t")
            if anchor_id in train_anchors:
                continue
            fout.write(line)
            
# query 
train_queries = set()
with open(os.path.join(in_dir, "train_queries.tsv")) as fin:
    for line in fin:
        qid, _ = line.strip().split("\t")
        train_queries.add(qid)

val_query_in_train = 0
with open(os.path.join(in_dir, "val_queries.tsv")) as fin:
    with open(os.path.join(out_dir, "val_queries.tsv"), "w") as fout:
        for line in fin:
            qid, _ = line.strip().split("\t")
            if qid in train_queries:
                val_query_in_train += 1
                continue
            fout.write(line)

with open(os.path.join(in_dir, "val_query_qrels.tsv")) as fin:
    with open(os.path.join(out_dir, "val_query_qrels.tsv"), "w") as fout:
        for line in fin:
            qid, _, _, _ = line.strip().split("\t")
            if qid not in train_queries:
                fout.write(line)
                
with open(os.path.join(in_dir, "val_query_qrels_clicknum.tsv")) as fin:
    with open(os.path.join(out_dir, "val_query_qrels_clicknum.tsv"), "w") as fout:
        for line in fin:
            qid, _, _ = line.strip().split("\t")
            if qid not in train_queries:
                fout.write(line)
                
# -----------------------
test_query_in_train = 0
with open(os.path.join(in_dir, "test_queries.tsv")) as fin:
    with open(os.path.join(out_dir, "test_queries.tsv"), "w") as fout:
        for line in fin:
            qid, _ = line.strip().split("\t")
            if qid in train_queries:
                test_query_in_train += 1
                continue
            fout.write(line)
            
with open(os.path.join(in_dir, "test_query_qrels.tsv")) as fin:
    with open(os.path.join(out_dir, "test_query_qrels.tsv"), "w") as fout:
        for line in fin:
            qid, _, _, _ = line.strip().split("\t")
            if qid not in train_queries:
                fout.write(line)
                
with open(os.path.join(in_dir, "test_query_qrels_clicknum.tsv")) as fin:
    with open(os.path.join(out_dir, "test_query_qrels_clicknum.tsv"), "w") as fout:
        for line in fin:
            qid, _, _ = line.strip().split("\t")
            if qid not in train_queries:
                fout.write(line)
```
