```python
import os 

root_dir = "/home/jupyter/unity_jointly_rec_and_search/datasets/amazon_esci_dataset/task_2_multiclass_product_classification/"
test_dir = os.path.join(root_dir, "unified_test")
out_dir = os.path.join(root_dir, "entity_prefix")
if not os.path.exists(out_dir):
    os.mkdir(out_dir)

def add_query_prefix(text):
    return "query: " + text

def add_product_prefix(text):
    return "product: " + text

pid_to_title = {}
with open(os.path.join(root_dir, "collection_title.tsv")) as fin:
    for line in fin:
        pid, title = line.strip().split("\t")
        pid_to_title[pid] = title
        
qid_to_query = {}
with open(os.path.join(root_dir, "all_queries.tsv")) as fin:
    for line in fin:
        qid, query = line.strip().split("\t")
        qid_to_query[qid] = query

# write
with open(os.path.join(out_dir, "prefix_collection_title.tsv"), "w") as fout:
    for pid, title in pid_to_title.items():
        title = add_product_prefix(title)
        fout.write(f"{pid}\t{title}\n")

with open(os.path.join(out_dir, "prefix_all_queries.tsv"), "w") as fout:
    for qid, query in qid_to_query.items():
        query = add_query_prefix(query)
        fout.write(f"{qid}\t{query}\n")
        
with open(os.path.join(out_dir, "prefix_all_entities.tsv"), "w") as fout:
    for pid, title in pid_to_title.items():
        title = add_product_prefix(title)
        fout.write(f"{pid}\t{title}\n")
    for qid, query in qid_to_query.items():
        query = add_query_prefix(query)
        fout.write(f"{qid}\t{query}\n")
        
with open(os.path.join(test_dir, "anchors.test.sim.small.tsv")) as fin:
    with open(os.path.join(test_dir, "prefix_anchors.test.sim.small.tsv"), "w") as fout:
        for line in fin:
            array = line.strip().split("\t")
            assert len(array) == 3
            pid, title, relation = array
            title = add_product_prefix(title)
            fout.write(f"{pid}\t{title}\t{relation}\n")
            
with open(os.path.join(test_dir, "anchors.test.compl.tsv")) as fin:
    with open(os.path.join(test_dir, "prefix_anchors.test.compl.tsv"), "w") as fout:
        for line in fin:
            array = line.strip().split("\t")
            assert len(array) == 3
            pid, title, relation = array
            title = add_product_prefix(title)
            fout.write(f"{pid}\t{title}\t{relation}\n")
            
with open(os.path.join(test_dir, "queries.test.tsv")) as fin:
    with open(os.path.join(test_dir, "prefix_queries.test.tsv"), "w") as fout:
        for line in fin:
            array = line.strip().split("\t")
            assert len(array) == 3
            qid, query, relation = array
            query = add_query_prefix(query)
            fout.write(f"{qid}\t{query}\t{relation}\n")
```
